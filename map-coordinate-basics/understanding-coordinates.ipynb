## **Information of the coordinate systems**

- By checking the **`Layer Properties`**, we can obtain the following info about the coordinate system used in map


| Variable      | Information  |
| ----------------------------- | -------------------------- |
| Projection used in map     | We are warned of specific distortion caused by projection |
| Projection zone number     | The zone number of a specific projection tells us which region of the Earth it is best suited for, that is, where on Earth is least distorted in all respects   |
| Coordinate system      | How data will overlap with data/how the data is stored   |
| Datum | What/where is the reference point for all the data in the coordinate system, where are measurements made from |
